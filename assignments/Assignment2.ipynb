```python
# we start by importing the packages that will be used
import calendar
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import RandomizedSearchCV
```

```python
df= pd.read_csv('Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv')
```

<font color='green'> We will work with **BURGLARY** and **FRAUD** due to the fact that there are many crimes of this types.
</font>

```python
crimes=set(['BURGLARY', 'FRAUD'])

# we get the 168 hours of the week
df['Time'] = [x + ':00' for x in df['Time']]
df['TimeInt']=pd.to_timedelta(df.Time, unit='hours').dt.floor('H')
df['HourInt']=  df['TimeInt']/ np.timedelta64(1, 'h')

days = dict(zip(calendar.day_name, range(7))); 
df['Day_nr'] = df['DayOfWeek'].map(lambda x: days[x])
# hour 0 corresponds to monday 00:00 - 01:00
df['Interval168']=df['Day_nr']*24+df['HourInt']

# df_rf corresponds to the dataframe that will be used on the random forest classifier
df_rf=df[df.Category.isin(crimes)]
```

```python
# we can check how many of each category of crime do we have and also drop NaNs
df_rf=df_rf.dropna()
df_rf['Category'].value_counts()
```

<font color='green'> Now, it is time to encode the attribute `PdDistrict`, first we will use a scikit's LabelEncoder, which will transform the `string` into an integer, so that we can pass it to our random forest classifier.
    
Later on we will do the transformation using one-hot encoding and we'll compare the performance of our classifier. What do you think will happen? Let's find out ðŸ˜€
</font>

```python
le = preprocessing.LabelEncoder()
le.fit(df_rf['PdDistrict'])
df_rf['PdDistrict_nr'] = le.transform(df_rf['PdDistrict'])
```

```python
'''Here we select the features that we want to use on the classifier
which are the 168 hours of the week and the district.'''
df_X1=df_rf[['Interval168','PdDistrict_nr']]
df_y1=df_rf['Category']

#then we transform it into a numpy array, since that is what the classifier "receives"
X1=df_X1.to_numpy()
y1=df_y1.to_numpy()

''' Separation of the dataset between training and test data.
the test size is 33% of the whole dataset.'''
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.33, random_state=42)

unique_elements, counts_elements = np.unique(y1_train, return_counts=True)

print("Frequency of values on the train dataset")
print(np.asarray((unique_elements, counts_elements)))

#we build the random forest
rf1 = RandomForestClassifier(n_estimators=100,random_state = 42,class_weight='balanced')

rf1.fit(X1_train,y1_train)
y1_pred=rf1.predict(X1_test)

print('Accuracy score:')
print(accuracy_score(y1_test, y1_pred))
print('Classification report on test data:')
print(classification_report(y1_test, y1_pred))
print('Confusion Matrix:')
print(confusion_matrix(y1_test, y1_pred))
print('Classification report on train data (check for overfitting):')
print(classification_report(rf1.predict(X1_train), y1_train))


# we can verify the precision with a function defined by us as well
def myprecision(y,ypred, classlist):
    TP,FP=[0,0],[0,0]
    for i in range(len(y)):
        if ypred[i]==y[i]:
            if ypred[i]==classlist[0]:
                TP[0]+=1
            else:
                TP[1]+=1
        else:
            if ypred[i]==classlist[0]:
                FP[0]+=1
            else:
                FP[1]+=1
    return [classlist[i]+': '+str(TP[i]/(FP[i]+TP[i])) for i in range(len(TP))]

myprecision(y1_test,y1_pred,['BURGLARY','FRAUD'])
```

<font color='green'> So it is time to check how the model performs with one-hot encoding by using pandas get_dummies function.
</font>

```python
df_rf = pd.concat([df_rf,pd.get_dummies(df_rf['PdDistrict'], prefix='PdDistrict')],axis=1)
list(df_rf.columns)
```

```python
'''Here we select the features that we want to use on the classifier
which are the 168 hours of the week and the district.'''
df_X2=df_rf[['Interval168', 'PdDistrict_BAYVIEW','PdDistrict_CENTRAL','PdDistrict_INGLESIDE','PdDistrict_MISSION','PdDistrict_NORTHERN',
             'PdDistrict_PARK','PdDistrict_RICHMOND','PdDistrict_SOUTHERN','PdDistrict_TARAVAL','PdDistrict_TENDERLOIN']]
df_y2=df_rf['Category']


X2=df_X2.to_numpy()
y2=df_y2.to_numpy()


X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.33, random_state=42)

#we build the random forest
rf2 = RandomForestClassifier(n_estimators=100,random_state = 42,class_weight='balanced')

rf2.fit(X2_train,y2_train)
y2_pred=rf2.predict(X2_test)

print(accuracy_score(y2_test, y2_pred))
print(classification_report(y2_test, y2_pred))
print(confusion_matrix(y2_test, y2_pred))
```

<font color='green'> So, the random forest classifier give us **exactly** the same results when using LabelEncoder and one-hot encoding. The explanation lies in that RandomForestClassifier does not handles the values on PdDistrict_nr in the first classifier to be **x** times as good a as another one (if PdDistrict_nr is 1 or 9 it does not affect the classification). It just finds the best splits to minimize entropy.
    
Now, we are going to use CV to determine the best model features.
</font>

```python
from pprint import pprint
# Look at parameters used by our current forest
print('Parameters currently in use:\n')
pprint(rf1.get_params())
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 500, num = 50)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 80, num = 5)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
pprint(random_grid)
```

```python
'''Careful here, this takes around 8 minutes to run,
so either trust our output or grab a coffee and wait for the results :)'''
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rfcv = RandomForestClassifier(class_weight='balanced')
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rfcv, param_distributions = random_grid, n_iter = 50, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(X1_train, y1_train)
```

```python
rf_random.best_params_
```

```python
best_random = rf_random.best_estimator_

y_pred_cv=best_random.predict(X1_test)


print(accuracy_score(y1_test, y_pred_cv))
print(classification_report(y1_test, y_pred_cv))
print(confusion_matrix(y1_test, y_pred_cv))
myprecision(y1_test,y_pred_cv,['BURGLARY','FRAUD'])
```

<font color='green'> We can see that tuning the parameters of the model by using cross validation improves the performance of the model against previously unseen data (test set). But, the improvement not a big one. we can see that the precision changes from:['BURGLARY: 0.7874637767456631', 'FRAUD: 0.44526911576249467'] to ['BURGLARY: 0.7878907639410346', 'FRAUD: 0.44724967769660506'] which is an improvement of 0.054% and 0.44% of the prediction of BURGLARY and FRAUD, respectively.


We kept this model simple as we want to see if the weather is actually correlated with the occurrence of crimes of type 1. We chose to balance the training data using `class_weight='balanced'`,  the reason is that there is almost double of frequency of BURGLARY compared to FRAUD, meaning that the classifier would lean towards classifyng according to the most repeated category, rather than centering in the features to classify the data. By doing this, it is expected that the misclassifications are somewhat similar between the two type of crimes.

It appears that the model does not overfit, we can check it by comparinf the classification report  on the train set and the test set. The performance of the model is not considerable different between them, which suggest that the model is not overfitting and learning the data instead of learnin _from_ the data.

As we explained before, we do get better results when using cross-validation at expense of the model taking more time to run.

```python
for i in focuscrimes:
    total = crimes_new[i].sum()
    crimes_new[i] = crimes_new[i].apply(lambda x: x/total)
crimes_new.head()# Veiwing that it is done right
```

#### 2. Creating the Bokeh plot.  
First a Bokeh dataframe is created from the calculated dataframe above 

```python
source = ColumnDataSource(crimes_new)
```

The hours are converted from the dataframe are converted to a list,  and the datatype is a string, since the FactorRange operates with this datatype.

```python
hours = [] # empty list to store the string converted hours
for i in range(24):
    hours.append(str(i))
```

The figure is created. Here the list created above is used to define the range of the x axis. The size of the plot is defined, and the title is added here.

```python
p = figure(x_range = FactorRange(factors=hours),plot_height=500, plot_width= 1200, toolbar_location=None, title="Crimes")
```

The dictionary "bar" and the list "items" are created to store the information, derived from the following for loop, and then used to plot the information. Pallette Category20 has been used for the color scheme of the different categories. The muted_color has the same colors as the filled color. Others When untoggled, it will show a blue color, which can be a litte frustrating to look at.  

```python
bar ={}
items = [] 
for indx,i in enumerate(focuscrimes):
    bar[i] = p.vbar(x="Hour", width=0.5,  top=i, source = source, muted_alpha = 0.06, alpha=3, muted_color=Category20[14][indx] ,color=Category20[14][indx]) 
    items.append((i, [bar[i]])) #the dictionary is appended to the items list, to the respective category.  
```

Additional information regarding the plot is added. This is the Axis titles, pushing the bars to the right, so they do not collide with the axis, but stand freely. 

```python
legend = Legend(items=items, location=(0,50)) #Legend and location is configured
p.add_layout(legend, "left") #Adding the legend outisde the plot. 
p.xaxis.axis_label = "Hours of the day" #Setting x-axis title
p.yaxis.axis_label = "Normalized crimes pr hour" #Setting y-axis title
p.x_range.range_padding = 0.05 # 
p.legend.click_policy="hide" #mute is chossen, becuase this enables one to see all of the different categories partily if not chosen, instead of hiding them. 

show(p) #shows the plot
```

From the plot above we can see how the different crimes distribute according to each other in a neat way.
