## Creating the same plot as above with comments


#### 1. Data is fetched and proccessed
Fetching the crime data and previewing the first lines of the dataframe

```python
crime = pd.read_csv("Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv", sep=",")
crime.head()
```

A column with the hour from the "Time" column is made. The Dataframe is also reduced to only obtain the focuscrimes. 

```python
crime["Hour"] = pd.to_datetime(crime["Time"], format = "%H:%M").dt.hour # hour cloumn is made. 
focuscrimes = ['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT']
crime = crime[crime["Category"].isin(focuscrimes)] #Reducing the dataframe only to obtain the forcuscrimes
```

The datafame is grouped by the hours coloumn, this creates a new dataframe with the counts, of the different crimes, according to the hours of the day. Veiwing the first line, gives a good indication of what is happening. **Hour 24 is hour 0 in our data**.

```python
crimes_new = crime.groupby('Hour')['Category'].value_counts().unstack()
crimes_new.reset_index(level=0, inplace=True) # creating a new index, so hour becomes a column. 
crimes_new.head()
```

Calculating the normalized arrests pr hour. As we can see from the first lines, they are close the the small preview in the text. Therefor we can assume it is done right.

```python
for i in focuscrimes:
    crimes_new[i] = crimes_new[i].apply(lambda x: x/crimes_new[i].sum())
crimes_new.head()# Veiwing that it is done right
```

#### 2. Creating the Bokeh plot.  
First a Bokeh dataframe is created from the calculated dataframe above 

```python
source = ColumnDataSource(crimes_new)
```

The hours are convertes from the dataframe are converted to a list,  and the datatype is s string, since the FactorRange operates with this datatype.

```python
hours = [] # empty list to store the string converted hours
for i in range(24):
    hours.append(str(i))
```

The figure is created. Here the list created above is used to define the range of the x axis. The size of the plot is defined, and the title is added here.

```python
p = figure(x_range = FactorRange(factors=hours),plot_height=500, plot_width= 1200, toolbar_location=None, title="Crimes")
```

The dictionary "bar" and the list "items" are created to store the information, derived from the following for loop, and then used to plot the information. Pallette Category20 has been used for the color scheme of the different categories. The muted_color has the same colors as the filled color. Others When untoggled, it will show a blue color, which can be a litte frustrating to look at.  

```python
bar ={}
items = [] 
for indx,i in enumerate(focuscrimes):
    bar[i] = p.vbar(x="Hour", width=0.5,  top=i, source = source, muted_alpha = 0.06, alpha=3, muted_color=Category20[14][indx] ,color=Category20[14][indx]) 
    items.append((i, [bar[i]])) #the dictionary is appended to the items list, to the respective category.  
    legend = Legend(items=items, location=(0,50)) #Legend and location is configured
```

Additional information regarding the plot is added. This is the Axis titles, pushing the bars to the right, so they do not collide with the axis, but stand freely. 

```python
p.add_layout(legend, "left") #Adding the legend outisde the plot. 
p.xaxis.axis_label = "Hours of the day" #Setting x-axis title
p.yaxis.axis_label = "Normalized crimes pr hour" #Setting y-axis title
p.x_range.range_padding = 0.05 # 
p.legend.click_policy="mute" #mute is chossen, becuase this enables one to see 

show(p) #shows the plot
```

From the plot above we can see how the different crimes distribute according to each other in a neat way.
