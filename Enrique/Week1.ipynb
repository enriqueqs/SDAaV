```python
df.Date=pd.to_datetime(df.Date, format='%m/%d/%Y', errors='coerce')
```

```python
df.groupby(df['Date'].map(lambda x: x.year)).IncidntNum.count()[:-1].mean()
```

```python
focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])
```

```python
df['Year'] = df['Date'].map(lambda x: x.year)
df=df[df.Year<2018]
```

```python
foc_crimes_year=df[df.Category.isin(focuscrimes)].groupby(['Year','Category'])['Category'].count()
foc_crimes_year
```

```python
#df_foc_crimes=pd.DataFrame(foc_crimes_year)
#df_foc_crimes.unstack().plot.bar(subplots=True,figsize=(10,60), legend=False)
```

```python
a=1

plt.figure(figsize=(20,15))

for i in focuscrimes:
    plt.subplot(7,2,a)
    plt.bar(foc_crimes_year[:,i].index ,foc_crimes_year[:,i], label=i)
    a+=1
plt.show()
```
